## 1.
Write a python script to fetch news from the web API given below and show news headlines to user.

API details:

API Endpoint(URL) : https://gnewsapi.herokuapp.com

Output format : JSON


```python
import requests
from pprint import pprint
```

```python
url = 'https://gnewsapi.herokuapp.com/'
reql = requests.get(url)
```

```python
data = reql.json()
#print(pprint(data))
```

```python
for i in range(len(data)):
    headline = data[i]['title']
    pprint(headline)
```
